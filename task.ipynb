```python
class Patient:
    def __init__(self, PId, Pname, Page, Pdisease):
        self.P_Id = PId
        self.P_Name = Pname
        self.P_Age = Page
        self.P_Disease = Pdisease

    def get_patient_details(self):
        return {
            'Patient_Id': self.P_Id,
            'Patient_Name': self.P_Name,
            'Patient_Age': self.P_Age,
            'Patient_Disease': self.P_Disease
        }


class Doctor(Patient):
    def __init__(self, doctor_id, name, specialization, PId, Pname, Page, Pdisease):
        super().__init__(PId, Pname, Page, Pdisease)
        self.doctor_id = doctor_id
        self.name = name
        self.specialization = specialization

    def get_doctor_details(self):
        return {
            'Doctor_Id': self.doctor_id,
            'Name': self.name,
            'Specialization': self.specialization
        }


class Appointment(Doctor):
    def __init__(self, appointment_id, patient_name, doctor_name, date,
                 doctor_id, name, specialization, PId, Pname, Page, Pdisease):
        super().__init__(doctor_id, name, specialization, PId, Pname, Page, Pdisease)
        self.appointment_id = appointment_id
        self.patient_name = patient_name
        self.doctor_name = doctor_name
        self.date = date

    def get_appointment_details(self):
        return {
            'Appointment_Id': self.appointment_id,
            'Patient_Name': self.patient_name,
            'Doctor_Name': self.doctor_name,
            'Date': self.date
        }


class Hospital(Appointment):
    def __init__(self, HId, Hname, Hlocation,
                 appointment_id, patient_name, doctor_name, date,
                 doctor_id, name, specialization, PId, Pname, Page, Pdisease):
        super().__init__(appointment_id, patient_name, doctor_name, date,
                         doctor_id, name, specialization, PId, Pname, Page, Pdisease)
        self.HId = HId
        self.Hname = Hname
        self.Hlocation = Hlocation

    def get_hospital_details(self):
        return {
            'Hospital_Id': self.HId,
            'Hospital_Name': self.Hname,
            'Hospital_Location': self.Hlocation
        }
```
